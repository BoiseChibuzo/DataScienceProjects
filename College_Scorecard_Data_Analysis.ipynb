## Quest 3: Distribution of Admission rate and  buckectization using a computed Median

<em><blockquote>Here, i used the Adm_rate variable instead of Adm_rate_all because Adm_rate variable is the individual campuses admission rate, and this brings out more insights into the data than the Adm_rate_all that cuts across schools which can shield some insights from the data.</blockquote><em>

```python
#Numerically described.
schools['ADM_RATE'].describe()#To list the statistics-mean,median,std etc of the distribution
```

```python
#visually describing the distribution of Adm_rate continuous variable
plt.figure(figsize=(18,9))# size of the plot area
plt.title('The visual Description of the Adm_rate Continuous variable:The graph is skewed to the left;more observations to the right and fewer to the left',fontsize='16')
plt.hist(schools['ADM_RATE'],bins=20,color='g')
plt.xlabel('Admission Rate',fontsize='16')
plt.ylabel('Values',fontsize='16')
plt.show()
```

<em><blockquote>Calculating the median to be used to bucketize the Admission rate by Admission Policy</blockquote><em>

```python
# Calculating the median and stroing in the variable-buck_med
buck_med =np.median(schools['ADM_RATE'].dropna())
buck_med #Median
```

```python
#To create AdmissionPolicy index to be same as that of the dataframe-schools
AdmissionPolicy = pd.Series('Open Admission', index=schools.index)
```

<em><blockquote>Comparing the Adm_rate with the median in order to finally group them into the specified groups;If Adm_rate is greater than the median then "Low Selectivity bucket",If Adm_rate is less than the computed median you have "High Selectivity", and others categorized into open admission</blockquote><em>

```python
#If Adm_rate is greater than the median then it is low selectivity
AdmissionPolicy[schools.ADM_RATE > buck_med] = 'Low Selectivity'
```

..

```python
#If Adm_rate is less than the median then it is low selectivity
AdmissionPolicy[schools.ADM_RATE < buck_med] = 'High Selectivity' #Using the Median to bucketize the Adm_rate
```

<em><blockquote>Displaying the bucketized Admission policy series</blockquote><em>

```python
AdmissionPolicy #Display the Series AdmissionPolicy
```

```python
#Adding this AdmissionPolicy variable to the dataframe
schools['AdmissionPolicy'] = AdmissionPolicy
```

<em><blockquote>After adding back the AdmissionPolicy series into the Dataframe-schools,i have to display the dataframe to see if it was added </blockquote><em>

```python
schools # Display the Dataframe- schools to see if the AdmissionPolicy has been added
```

```python
#Grouping by AdmissionPolicy to show as a table
hist1=schools.groupby('AdmissionPolicy')['INSTNM'].value_counts()
hist1
```

```python
#Displaying the distribution using seaborn's countplot

plt.figure(figsize=(9,7))
sns.countplot(AdmissionPolicy)
sns.set_theme(style="darkgrid")
```

```python
#Convert AdmissionPolicy to a categorical variable
AdmissionPolicy = AdmissionPolicy.astype('category')
```

```python
#Reordering the variable so the highest will be the first bar
AdmissionPolicy = AdmissionPolicy.cat.reorder_categories(['Open Admission', 'Low Selectivity', 'High Selectivity'])
```

```python
schools['AdmissionPolicy'] = AdmissionPolicy # Adding the AdmissionPolicy variable back to the dataframe
```

```python
#Plotting to show the bar chart after the re ordering
plt.figure(figsize=(9,7))#
plt.title('Fig:Admission Policy Categorization of Schools',fontsize=14)
plt.xlabel('Admission Policy')
plt.ylabel('Number of Schools in each category')
sns.set_theme(style="darkgrid")
catbar=sns.countplot(AdmissionPolicy)

```

## Question 4: Dissagregation by Race:


<em><blockquote>Here, i selected C150_4_WHITE and C150_L4_*, i selected the first 9 columns for C150_4_* and 9columns of C150_L4_* differently, and changed the names of the L4s to enable me combine the two nine columns. After that, combine_first was used to combine to arrive at one dataframe.</blockquote><em>

```python
#This is the dataframe of 4yrs races
race4=schools[['C150_4_WHITE','C150_4_BLACK','C150_4_HISP','C150_4_ASIAN','C150_4_AIAN','C150_4_NHPI','C150_4_2MOR','C150_4_NRA','C150_4_UNKN'
]]
race4
```

```python
race_l4=schools[['C150_L4_WHITE','C150_L4_BLACK','C150_L4_HISP','C150_L4_ASIAN','C150_L4_AIAN','C150_L4_NHPI','C150_L4_2MOR','C150_L4_NRA','C150_L4_UNKN'
]]
race_l4
# This is the dataframe of Less_than_four years races
```

```python
#changing the Less than 4yrs names to that of 4yrs to make it easier to combine by using the set_axis
#the next thing would be to use the combine_first function to fill in the missing values
race_l4new=race_l4.set_axis(list(race4.columns),axis=1)
race_l4new
```

```python
#Using the combine_first function to combine 4yr and less than 4yr completion rate variables
Compl_by_race=race4.combine_first(race_l4new).dropna()
Compl_by_race
#Combining to fill in the missing values
```

<em><blockquote>Below is a code to change the name of the race  variables to a clearer understandable names</blockquote></em>

```python
#Renaming the Variables with understandable names of races
Compl_by_race = Compl_by_race.rename(columns={'C150_4_WHITE':'WHITE', 'C150_4_BLACK':'BLACK',
                                                 'C150_4_HISP':'HISPANIC','C150_4_ASIAN':'ASIAN','C150_4_AIAN':'AIAN','C150_4_NHPI':'NHPI', 
                                                 'C150_4_2MOR':'2MOR', 'C150_4_NRA':'NRA', 'C150_4_UNKN':'UNKN'})
Compl_by_race
```

<em><blockquote>Describing numerically: </blockquote><em>

```python
#Describing numerically to show the mean and other statistic across the races
Compl_by_race.describe()
```

<em><blockquote>Melting the dataframe to a long format and renaming the variables for easy plotting: </blockquote><em>

```python
#Transforming to a long format to enable plotting the distribution using the melt function and changing the variable and value name to something meaningful
Dist_Comp_by_race=Compl_by_race.melt(var_name='Race',value_name='Completion Rate')
Dist_Comp_by_race
```

<em><blockquote>Describing Visually: </blockquote><em>

```python
# Describing the distribution visually

plt.figure(figsize=(15,10))# Set the size of the plot figure.
plt.title('Fig:The Completion rate by race of 4yr and Less than 4yr shools,NHPI race is eye-catchy here, no outliers',fontsize=14)
plt.xlabel('RACE',fontsize=14)
plt.ylabel('COMPLETION RATE',fontsize=14)
sns.boxplot(x='Race',y='Completion Rate',data=Dist_Comp_by_race)
```
